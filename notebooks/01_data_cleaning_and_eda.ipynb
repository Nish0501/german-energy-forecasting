```python
# Load the raw data using the header parameter to read the correct column names
file_path = 'G:/MyProjects/german_energy_forecasting/data/raw/TSO_actual.csv'
df = pd.read_csv(file_path, header=[0, 1], index_col=0)

# Display the first 5 rows and a summary of the data
print("First 5 rows after setting header:")
print(df.head())
print("\nDataFrame Info:")
df.info()
```

```python
# Select only the columns related to energy generation and demand
energy_cols = [col for col in df.columns if col not in ['Unnamed: 0_level_0_Unnamed: 0_level_1', 'DateTime_..._...']]

# Create a new DataFrame with a single column for each energy type
df_aggregated = df[energy_cols].copy()

# Use str.split to get the first part of the column name (the energy type)
df_aggregated.columns = df_aggregated.columns.str.split('_').str[0]

# Group by the new column names and sum across columns (axis=1)
df_final = df_aggregated.groupby(level=0, axis=1).sum()

# Display the new columns and the first few rows
print("Final aggregated columns:")
print(df_final.columns)
print("\nFirst 5 rows of the final DataFrame:")
print(df_final.head())
```

```python
# Rename the columns to be more descriptive
new_column_names = {
    'B': 'Biomass',
    'D': 'Demand',
    'F': 'Fossil',
    'N': 'Nuclear',
    'S': 'Solar',
    'W': 'Wind'
}
df_final = df_final.rename(columns=new_column_names)

# Display the final columns and data
print("Final aggregated columns:")
print(df_final.columns)
print("\nFirst 5 rows of the final DataFrame:")
print(df_final.head())
```

```python
# Resample the data to a daily frequency and sum the values
df_daily = df_final.resample('D').sum()

# Display the first 5 rows of the daily data
print("First 5 rows of the daily DataFrame:")
print(df_daily.head())
```
